```python
import tkinter as tk
from tkinter import messagebox

library_books = []

def add_book():
    title = title_entry.get()
    author = author_entry.get()
    year = year_entry.get()
    if title and author and year:
        book = {"title": title, "author": author, "year": year}
        library_books.append(book)
        update_book_list()
        title_entry.delete(0, tk.END)
        author_entry.delete(0, tk.END)
        year_entry.delete(0, tk.END)
    else:
        messagebox.showerror("Error", "Enter all fields")

def delete_book():
    selected = book_list.curselection()
    if selected:
        index = selected[0]
        library_books.pop(index)
        update_book_list()
    else:
        messagebox.showerror("Error", "Select a book to delete")

def update_book_list():
    book_list.delete(0, tk.END)
    for book in library_books:
        display = f"{book['title']} by {book['author']} ({book['year']})"
        book_list.insert(tk.END, display)

window = tk.Tk()
window.title("Library Management System")
window.geometry("400x450")
window.resizable(False, False)

title_label = tk.Label(window, text="Book Title")
title_label.pack()
title_entry = tk.Entry(window, width=40)
title_entry.pack()

author_label = tk.Label(window, text="Author")
author_label.pack()
author_entry = tk.Entry(window, width=40)
author_entry.pack()

year_label = tk.Label(window, text="Year")
year_label.pack()
year_entry = tk.Entry(window, width=40)
year_entry.pack()

add_button = tk.Button(window, text="Add Book", width=30, command=add_book)
add_button.pack(pady=10)

delete_button = tk.Button(window, text="Delete Selected Book", width=30, command=delete_book)
delete_button.pack(pady=5)

book_list = tk.Listbox(window, width=50, height=10)
book_list.pack(pady=10)

window.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def add_student():
    name = name_entry.get()
    roll = roll_entry.get()
    course = course_entry.get()
    if name and roll and course:
        with open("students.txt", "a") as file:
            file.write(f"{name},{roll},{course}\n")
        name_entry.delete(0, tk.END)
        roll_entry.delete(0, tk.END)
        course_entry.delete(0, tk.END)
        messagebox.showinfo("Success", "Student added")
    else:
        messagebox.showerror("Error", "Enter all fields")

def view_students():
    student_list.delete(0, tk.END)
    try:
        with open("students.txt", "r") as file:
            for line in file:
                student_list.insert(tk.END, line.strip())
    except FileNotFoundError:
        open("students.txt", "w").close()

def search_student():
    keyword = search_entry.get().lower()
    student_list.delete(0, tk.END)
    found = False
    try:
        with open("students.txt", "r") as file:
            for line in file:
                if keyword in line.lower():
                    student_list.insert(tk.END, line.strip())
                    found = True
        if not found:
            messagebox.showinfo("Not Found", "No student matched")
    except FileNotFoundError:
        open("students.txt", "w").close()

def delete_student():
    name_to_delete = delete_entry.get().lower()
    lines = []
    deleted = False
    try:
        with open("students.txt", "r") as file:
            lines = file.readlines()
        with open("students.txt", "w") as file:
            for line in lines:
                if name_to_delete not in line.lower():
                    file.write(line)
                else:
                    deleted = True
        if deleted:
            messagebox.showinfo("Deleted", "Student record deleted")
        else:
            messagebox.showinfo("Not Found", "No student found")
        delete_entry.delete(0, tk.END)
        view_students()
    except FileNotFoundError:
        open("students.txt", "w").close()

window = tk.Tk()
window.title("Student Record Management System")
window.geometry("500x550")
window.resizable(False, False)

tk.Label(window, text="Name").pack()
name_entry = tk.Entry(window, width=40)
name_entry.pack()

tk.Label(window, text="Roll No").pack()
roll_entry = tk.Entry(window, width=40)
roll_entry.pack()

tk.Label(window, text="Course").pack()
course_entry = tk.Entry(window, width=40)
course_entry.pack()

tk.Button(window, text="Add Student", width=30, command=add_student).pack(pady=10)
tk.Button(window, text="View All Students", width=30, command=view_students).pack(pady=5)

student_list = tk.Listbox(window, width=60, height=10)
student_list.pack(pady=10)

tk.Label(window, text="Search by Name").pack()
search_entry = tk.Entry(window, width=30)
search_entry.pack()
tk.Button(window, text="Search", command=search_student).pack(pady=5)

tk.Label(window, text="Delete by Name").pack()
delete_entry = tk.Entry(window, width=30)
delete_entry.pack()
tk.Button(window, text="Delete", command=delete_student).pack(pady=5)

window.mainloop()
```
